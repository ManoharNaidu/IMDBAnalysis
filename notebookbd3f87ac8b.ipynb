# 20. Classify Movies Based on Ratings [Good,Better and Best]

```python
def rating(rating):
    if rating>=8.0: # changed 7.0 -> 8.0
        return 'Excellent'
    elif rating>=6.5: # changed 6.0 -> 6.5
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(rating)
data
```
